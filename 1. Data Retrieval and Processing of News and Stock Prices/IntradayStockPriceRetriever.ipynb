This python notebook imports intraday prices every hour in year 2024 and calculates the price change at every hour

```python
import pandas as pd
```

```python
df_eth= pd.read_csv('us_2024_ticker combined versions/ETHUSDT_Combined.csv')
df_eth['PriceChange']=((df_eth['<CLOSE>']-df_eth['<OPEN>'])/df_eth['<OPEN>'])*100
df_eth.to_csv('us_2024_ticker combined versions/ETHUSDT_Combined_WithPriceChange.csv')
```

```python
df_qqq = pd.read_csv('us_2024_ticker combined versions/QQQ_Combined.csv')
df_qqq['PriceChange']=((df_qqq['<CLOSE>']-df_qqq['<OPEN>'])/df_qqq['<OPEN>'])*100
df_qqq.to_csv('us_2024_ticker combined versions/QQQ_Combined_WithPriceChange.csv')
```

```python
df_spy= pd.read_csv('us_2024_ticker combined versions/SPY_Combined.csv')
df_spy['PriceChange']=((df_spy['<CLOSE>']-df_spy['<OPEN>'])/df_spy['<OPEN>'])*100
df_spy.to_csv('us_2024_ticker combined versions/SPY_Combined_WithPriceChange.csv')
```
